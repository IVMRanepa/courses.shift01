## üêç –ß–∞—Å—Ç—å 2: PEP8 - –°—Ç–∞–Ω–¥–∞—Ä—Ç –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è –∫–æ–¥–∞ Python

**PEP8** (Python Enhancement Proposal 8) ‚Äî —ç—Ç–æ —Ä—É–∫–æ–≤–æ–¥—Å—Ç–≤–æ –ø–æ —Å—Ç–∏–ª—é –∫–æ–¥–∞ Python.

### –û—Å–Ω–æ–≤–Ω—ã–µ –ø—Ä–∞–≤–∏–ª–∞ PEP8:
- **–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏:** 79 —Å–∏–º–≤–æ–ª–æ–≤ (–∫–æ–¥), 72 —Å–∏–º–≤–æ–ª–∞ (–∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏)
- **–û—Ç—Å—Ç—É–ø—ã:** 4 –ø—Ä–æ–±–µ–ª–∞ (–Ω–µ —Ç–∞–±—É–ª—è—Ü–∏—è)
- **–ü—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏:** –º–µ–∂–¥—É —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏ –∫–ª–∞—Å—Å–∞–º–∏ (2 —Å—Ç—Ä–æ–∫–∏), –≤–Ω—É—Ç—Ä–∏ —Ñ—É–Ω–∫—Ü–∏–π (1 —Å—Ç—Ä–æ–∫–∞)
- **–ò–º–ø–æ—Ä—Ç—ã:** –≤ –Ω–∞—á–∞–ª–µ —Ñ–∞–π–ª–∞, –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω—ã, –∞–ª—Ñ–∞–≤–∏—Ç–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫
- **–ü—Ä–æ–±–µ–ª—ã –≤–æ–∫—Ä—É–≥ –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤:** `x = 1 + 2` (–ø—Ä–∞–≤–∏–ª—å–Ω–æ), `x=1+2` (–Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ)


````
# ‚ùå –ü–õ–û–•–û–ô –ü–†–ò–ú–ï–†: –Ω–∞—Ä—É—à–µ–Ω–∏–µ PEP8
def calculate_total(prices,quantity):
    """–†–∞—Å—á–µ—Ç –æ–±—â–µ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏"""
    total=0
    for i in range(len(prices)):
        total=total+prices[i]*quantity[i]
    return total

result=calculate_total([10,20,30],[2,1,3])
print('–†–µ–∑—É–ª—å—Ç–∞—Ç:', result)
````

### ‚úÖ –ü—Ä–æ–±–ª–µ–º—ã –∫–æ–¥–∞ –≤—ã—à–µ:
1. –û—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –ø—Ä–æ–±–µ–ª–æ–≤ –≤–æ–∫—Ä—É–≥ –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤ `=` –∏ –≤ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö
2. –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ü–∏–∫–ª–∞ (–ª—É—á—à–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å `zip()`)
3. –ü–ª–æ—Ö–∏–µ –∏–º–µ–Ω–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö (`i` –≤–º–µ—Å—Ç–æ `index`)
4. –û—Ç—Å—É—Ç—Å—Ç–≤–∏–µ type hints

### ‚úÖ –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç:

````
# ‚úÖ –•–û–†–û–®–ò–ô –ü–†–ò–ú–ï–†: —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ PEP8
from typing import List

def calculate_total(prices: List[float],
                    quantity: List[int]) -> float:
    """
    –†–∞—Å—á–µ—Ç –æ–±—â–µ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏ —Ç–æ–≤–∞—Ä–æ–≤.

    Args:
        prices: –°–ø–∏—Å–æ–∫ —Ü–µ–Ω —Ç–æ–≤–∞—Ä–æ–≤
        quantity: –°–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤ —Ç–æ–≤–∞—Ä–æ–≤

    Returns:
        –û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤
    """
    # –í—ã—á–∏—Å–ª—è–µ–º —Å—É–º–º—É –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–π —Ü–µ–Ω—ã –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
    total = sum(
        price * qty
        for price, qty in zip(prices, quantity)
    )
    return total


# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
result = calculate_total([10, 20, 30], [2, 1, 3])
print(f"‚úÖ –û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å: {result} —Ä—É–±.")
print(f"‚úÖ –¢–∏–ø —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞: {type(result)}")
````

## üìã –ß–∞—Å—Ç—å 3: –ü—Ä–∏–Ω—Ü–∏–ø 4 - –°–æ–±–ª—é–¥–∞–π—Ç–µ –ø—Ä–∞–≤–∏–ª–∞ –∏–º–µ–Ω–æ–≤–∞–Ω–∏—è

**–ü—Ä–∞–≤–∏–ª–∞ –∏–º–µ–Ω–æ–≤–∞–Ω–∏—è (PEP8):**
- **–ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –∏ —Ñ—É–Ω–∫—Ü–∏–∏:** `snake_case` (lowercase_with_underscores)
- **–ö–ª–∞—Å—Å—ã:** `PascalCase` (CapitalizedWords)
- **–ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã:** `UPPERCASE_WITH_UNDERSCORES`
- **–ü—Ä–∏–≤–∞—Ç–Ω—ã–µ –º–µ—Ç–æ–¥—ã:** `_private_method` –∏–ª–∏ `__dunder_method__`

### ‚ùå –ü–ª–æ—Ö–∏–µ –∏–º–µ–Ω–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö:

````
# ‚ùå –ü–õ–û–•–û–ô –ü–†–ò–ú–ï–†: –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ

# –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∏–º–µ–Ω–∞
x = 1000  # –ß—Ç–æ —ç—Ç–æ? –ó–∞—Ä–ø–ª–∞—Ç–∞? –¶–µ–Ω–∞? –ù–µ—è—Å–Ω–æ!
y = ["Alice", "Bob", "Charlie"]  # –ö–∞–∫–æ–π —ç—Ç–æ —Å–ø–∏—Å–æ–∫?
fn = lambda a, b: a + b  # –ß—Ç–æ –≤—ã—á–∏—Å–ª—è–µ—Ç —ç—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è?
Var_Name = 42  # –°–º–µ—à–∏–≤–∞–Ω–∏–µ —Å—Ç–∏–ª–µ–π (–Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ)
FUNCTION_NAME = 123  # –ö–æ–Ω—Å—Ç–∞–Ω—Ç–∞ –∏–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è? –ó–∞–ø—É—Ç–∞–Ω–Ω–æ

print("‚ùå –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∏–º–µ–Ω–∞:")
print(f"x = {x}, y = {y}, fn(1,2) = {fn(1, 2)}")
````

### ‚úÖ –•–æ—Ä–æ—à–∏–µ –∏–º–µ–Ω–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö:

````
# ‚úÖ –•–û–†–û–®–ò–ô –ü–†–ò–ú–ï–†: –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ

# –ü—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∏–º–µ–Ω–∞ (PEP8)
monthly_salary = 1000  # –Ø—Å–Ω–æ, —á—Ç–æ —ç—Ç–æ –∑–∞—Ä–ø–ª–∞—Ç–∞
employee_names = ["Alice", "Bob", "Charlie"]  # –ü–æ–Ω—è—Ç–Ω–æ, —á—Ç–æ —Å–ø–∏—Å–æ–∫ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤

def add_numbers(a: int, b: int) -> int:
    """–°–∫–ª–∞–¥—ã–≤–∞–µ—Ç –¥–≤–∞ —á–∏—Å–ª–∞ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç."""
    return a + b

MAX_ATTEMPTS = 3  # –ö–æ–Ω—Å—Ç–∞–Ω—Ç–∞ (—à—Ä–∏—Ñ—Ç UPPERCASE)

class DataProcessor:
    """–ö–ª–∞—Å—Å –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö (PascalCase)."""

    def __init__(self):
        self._internal_state = None  # –ü—Ä–∏–≤–∞—Ç–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç

    def process_data(self, data: list) -> list:
        """–û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ (snake_case –¥–ª—è –º–µ—Ç–æ–¥–æ–≤)."""
        return [x * 2 for x in data]

print("‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∏–º–µ–Ω–∞:")
print(f"–ó–∞—Ä–ø–ª–∞—Ç–∞: {monthly_salary} —Ä—É–±.")
print(f"–°–æ—Ç—Ä—É–¥–Ω–∏–∫–∏: {employee_names}")
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {add_numbers(5, 3)}")
print(f"–ú–∞–∫—Å–∏–º—É–º –ø–æ–ø—ã—Ç–æ–∫: {MAX_ATTEMPTS}")
print(f"–û–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: {DataProcessor().process_data([1, 2, 3])}")
````

4. **–†–µ–∞–ª–∏–∑–∞—Ü–∏—è** (—á–µ—Ç–∫–∞—è, –ø—Ä–æ—Å—Ç–∞—è –ª–æ–≥–∏–∫–∞)
5. **–í–æ–∑–≤—Ä–∞—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞** (–∏–ª–∏ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ)


````
# ‚úÖ –•–û–†–û–®–ò–ô –ü–†–ò–ú–ï–†: —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ —Ñ—É–Ω–∫—Ü–∏–∏ —Å–æ–≥–ª–∞—Å–Ω–æ PEP8
from typing import Optional, Union
from datetime import datetime


def validate_email(email: str) -> bool:
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç –≤–∞–ª–∏–¥–Ω–æ—Å—Ç—å –∞–¥—Ä–µ—Å–∞ —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –ø–æ—á—Ç—ã.

    Args:
        email: –°—Ç—Ä–æ–∫–∞ —Å –∞–¥—Ä–µ—Å–æ–º —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –ø–æ—á—Ç—ã

    Returns:
        True –µ—Å–ª–∏ email –≤–∞–ª–∏–¥–Ω—ã–π, False –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ

    Raises:
        TypeError: –ï—Å–ª–∏ email –Ω–µ —Å—Ç—Ä–æ–∫–∞

    Examples:
        >>> validate_email("user@example.com")
        True
        >>> validate_email("invalid-email")
        False
    """
    if not isinstance(email, str):
        raise TypeError("Email –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Å—Ç—Ä–æ–∫–æ–π")

    # –ü—Ä–æ—Å—Ç–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ (–≤ —Ä–µ–∞–ª—å–Ω–æ—Å—Ç–∏ –ª—É—á—à–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å regex)
    return "@" in email and "." in email.split("@")[-1]


def calculate_discount(price: float,
                      discount_percent: float) -> float:
    """
    –†–∞—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ü–µ–Ω—É —Å —É—á–µ—Ç–æ–º —Å–∫–∏–¥–∫–∏.

    Args:
        price: –ò—Å—Ö–æ–¥–Ω–∞—è —Ü–µ–Ω–∞ (—Ä—É–±.)
        discount_percent: –ü—Ä–æ—Ü–µ–Ω—Ç —Å–∫–∏–¥–∫–∏ (0-100)

    Returns:
        –¶–µ–Ω–∞ –ø–æ—Å–ª–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è —Å–∫–∏–¥–∫–∏

    Raises:
        ValueError: –ï—Å–ª–∏ —Ü–µ–Ω–∞ –∏–ª–∏ —Å–∫–∏–¥–∫–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ
    """
    if price < 0:
        raise ValueError("–¶–µ–Ω–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π")
    if not (0 <= discount_percent <= 100):
        raise ValueError("–°–∫–∏–¥–∫–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –æ—Ç 0 –¥–æ 100%")

    return price * (1 - discount_percent / 100)


# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π
print("‚úÖ –ü—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π:")
print()
print(f"Email 'user@example.com' –≤–∞–ª–∏–¥–Ω—ã–π: {validate_email('user@example.com')}")
print(f"Email 'invalid' –≤–∞–ª–∏–¥–Ω—ã–π: {validate_email('invalid')}")
print()
original_price = 1000
discount = 20
final_price = calculate_discount(original_price, discount)
print(f"–ò—Å—Ö–æ–¥–Ω–∞—è —Ü–µ–Ω–∞: {original_price} —Ä—É–±.")
print(f"–°–∫–∏–¥–∫–∞: {discount}%")
print(f"–ò—Ç–æ–≥–æ–≤–∞—è —Ü–µ–Ω–∞: {final_price} —Ä—É–±.")
````

## üí¨ –ß–∞—Å—Ç—å 5: –ü—Ä–∏–Ω—Ü–∏–ø 6 - –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏

### –ü—Ä–∞–≤–∏–ª–∞ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤:
- **–ö–æ–º–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ –ü–û–ß–ï–ú–£, –∞ –Ω–µ –ß–¢–û** (–∫–æ–¥ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –ß–¢–û)
- **–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ç–æ—á–Ω—ã–º–∏ –∏ –∞–∫—Ç—É–∞–ª—å–Ω—ã–º–∏**
- **–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ docstrings –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–π –∏ –∫–ª–∞—Å—Å–æ–≤**
- **–ò–∑–±–µ–≥–∞–π—Ç–µ –æ—á–µ–≤–∏–¥–Ω—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤**


````
# ‚ùå –ü–õ–û–•–û–ô –ü–†–ò–ú–ï–†: –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏

def calculate_average_bad(numbers):
    """–§—É–Ω–∫—Ü–∏—è."""
    total = 0  # –ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è —Å—É–º–º—ã
    for num in numbers:  # –¶–∏–∫–ª –ø–æ —á–∏—Å–ª–∞–º
        total = total + num  # –î–æ–±–∞–≤–ª—è–µ–º —á–∏—Å–ª–æ
    return total / len(numbers)  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç


# ‚úÖ –•–û–†–û–®–ò–ô –ü–†–ò–ú–ï–†: –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏

def calculate_average_good(numbers: list) -> float:
    """
    –í—ã—á–∏—Å–ª—è–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å–ø–∏—Å–∫–∞ —á–∏—Å–µ–ª.

    –ò—Å–ø–æ–ª—å–∑—É–µ–º –≤—Å—Ç—Ä–æ–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é sum() –≤–º–µ—Å—Ç–æ —Ä—É—á–Ω–æ–≥–æ —Ü–∏–∫–ª–∞
    –¥–ª—è –ª—É—á—à–µ–π –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∏ —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏.
    """
    if not numbers:
        # –û–±—Ä–∞–±–æ—Ç–∫–∞ –≥—Ä–∞–Ω–∏—á–Ω–æ–≥–æ —Å–ª—É—á–∞—è: –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫
        return 0.0

    # –ò—Å–ø–æ–ª—å–∑—É–µ–º –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤–º–µ—Å—Ç–æ —Ä—É—á–Ω–æ–≥–æ —Ü–∏–∫–ª–∞
    return sum(numbers) / len(numbers)


print("‚úÖ –ü—Ä–∏–º–µ—Ä—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ:")
data = [10, 20, 30, 40, 50]
print(f"–î–∞–Ω–Ω—ã–µ: {data}")
print(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {calculate_average_good(data)}")
print(f"–°—Ä–µ–¥–Ω–µ–µ –ø—É—Å—Ç–æ–≥–æ —Å–ø–∏—Å–∫–∞: {calculate_average_good([])}")
````

## üîÑ –ß–∞—Å—Ç—å 6: –ü—Ä–∏–Ω—Ü–∏–ø 9 - –ù–µ –ø–æ–≤—Ç–æ—Ä—è–π—Ç–µ—Å—å (DRY - Don't Repeat Yourself)

**–ü—Ä–∞–≤–∏–ª–æ DRY:** –ö–∞–∂–¥–∞—è –µ–¥–∏–Ω–∏—Ü–∞ –∑–Ω–∞–Ω–∏—è –¥–æ–ª–∂–Ω–∞ –∏–º–µ—Ç—å –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –≤ —Å–∏—Å—Ç–µ–º–µ.

### –ü—Ä–æ–±–ª–µ–º—ã –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏—è –∫–æ–¥–∞:
- –°–ª–æ–∂–Ω–µ–µ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞—Ç—å
- –õ–µ–≥—á–µ –≤–Ω–µ—Å—Ç–∏ –æ—à–∏–±–∫—É
- –ó–∞–Ω–∏–º–∞–µ—Ç –±–æ–ª—å—à–µ –º–µ—Å—Ç–∞
- –ú–µ–Ω–µ–µ —á–∏—Ç–∞–µ–º–æ


````
# ‚ùå –ù–ê–†–£–®–ï–ù–ò–ï DRY: –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–π—Å—è –∫–æ–¥

def process_invalid():
    """–ü—Ä–∏–º–µ—Ä –Ω–∞—Ä—É—à–µ–Ω–∏—è DRY –ø—Ä–∏–Ω—Ü–∏–ø–∞."""
    print("‚ùå –ü–æ–≤—Ç–æ—Ä—è—é—â–∏–π—Å—è –∫–æ–¥:")

    # –°—Ç—É–¥–µ–Ω—Ç 1
    name1 = "–ê–ª–∏—Å–∞"
    grade1 = 95
    if grade1 >= 90:
        rating1 = "–û—Ç–ª–∏—á–Ω–æ"
    elif grade1 >= 80:
        rating1 = "–•–æ—Ä–æ—à–æ"
    elif grade1 >= 70:
        rating1 = "–£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"
    else:
        rating1 = "–ù–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"
    print(f"{name1}: {grade1} -> {rating1}")

    # –°—Ç—É–¥–µ–Ω—Ç 2 (–ø–æ—á—Ç–∏ –∏–¥–µ–Ω—Ç–∏—á–Ω—ã–π –∫–æ–¥!)
    name2 = "–ë–æ–±"
    grade2 = 78
    if grade2 >= 90:
        rating2 = "–û—Ç–ª–∏—á–Ω–æ"
    elif grade2 >= 80:
        rating2 = "–•–æ—Ä–æ—à–æ"
    elif grade2 >= 70:
        rating2 = "–£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"
    else:
        rating2 = "–ù–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"
    print(f"{name2}: {grade2} -> {rating2}")

    # –°—Ç—É–¥–µ–Ω—Ç 3 (–ø–æ–≤—Ç–æ—Ä –µ—â–µ —Ä–∞–∑!)
    name3 = "–í–∏–∫—Ç–æ—Ä"
    grade3 = 88
    if grade3 >= 90:
        rating3 = "–û—Ç–ª–∏—á–Ω–æ"
    elif grade3 >= 80:
        rating3 = "–•–æ—Ä–æ—à–æ"
    elif grade3 >= 70:
        rating3 = "–£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"
    else:
        rating3 = "–ù–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"
    print(f"{name3}: {grade3} -> {rating3}")


process_invalid()
````

````
# ‚úÖ –†–ï–®–ï–ù–ò–ï: —Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ DRY –ø—Ä–∏–Ω—Ü–∏–ø—É

def get_rating(grade: int) -> str:
    """
    –ü–æ–ª—É—á–∞–µ—Ç —Ä–µ–π—Ç–∏–Ω–≥ —Å—Ç—É–¥–µ–Ω—Ç–∞ –ø–æ –æ—Ü–µ–Ω–∫–µ.

    Args:
        grade: –ß–∏—Å–ª–æ–≤–∞—è –æ—Ü–µ–Ω–∫–∞ (0-100)

    Returns:
        –°–ª–æ–≤–µ—Å–Ω—ã–π —Ä–µ–π—Ç–∏–Ω–≥
    """
    if grade >= 90:
        return "–û—Ç–ª–∏—á–Ω–æ"
    elif grade >= 80:
        return "–•–æ—Ä–æ—à–æ"
    elif grade >= 70:
        return "–£–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"
    else:
        return "–ù–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ"


def process_students(students: list) -> None:
    """
    –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –∏ –≤—ã–≤–æ–¥–∏—Ç –∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–∏.

    Args:
        students: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (–∏–º—è, –æ—Ü–µ–Ω–∫–∞)
    """
    print("‚úÖ –û–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –∫–æ–¥ (DRY):")
    for name, grade in students:
        rating = get_rating(grade)
        print(f"{name}: {grade} -> {rating}")


# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
students_data = [
    ("–ê–ª–∏—Å–∞", 95),
    ("–ë–æ–±", 78),
    ("–í–∏–∫—Ç–æ—Ä", 88),
    ("–ì–∞–ª–∏–Ω–∞", 65),
    ("–î–º–∏—Ç—Ä–∏–π", 91)
]

process_students(students_data)
````

## üéØ –ß–∞—Å—Ç—å 7: –ü—Ä–∏–Ω—Ü–∏–ø 10 - –ï–¥–∏–Ω–∞—è –æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç—å (SRP)

**Single Responsibility Principle:**
- –ö–ª–∞—Å—Å –∏–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –∏–º–µ—Ç—å –æ–¥–Ω—É –ø—Ä–∏—á–∏–Ω—É –¥–ª—è –∏–∑–º–µ–Ω–µ–Ω–∏—è
- –û–¥–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è = –æ–¥–Ω–∞ –∑–∞–¥–∞—á–∞
- –û–±–ª–µ–≥—á–∞–µ—Ç —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –ø–æ–¥–¥–µ—Ä–∂–∫—É


````
# ‚ùå –ù–ê–†–£–®–ï–ù–ò–ï SRP: –∫–ª–∞—Å—Å —Å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç—å—é

class UserBad:
    """
    –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ: –∫–ª–∞—Å—Å –æ—Ç–≤–µ—á–∞–µ—Ç –∑–∞ —Å–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ!
    """
    def __init__(self, name: str, email: str):
        self.name = name
        self.email = email

    def validate_email(self) -> bool:
        """–ü—Ä–æ–≤–µ—Ä–∫–∞ email (–±–∏–∑–Ω–µ—Å-–ª–æ–≥–∏–∫–∞)"""
        return "@" in self.email

    def save_to_database(self) -> None:
        """–°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ –ë–î (—Ä–∞–±–æ—Ç–∞ —Å –¥–∞–Ω–Ω—ã–º–∏)"""
        print(f"‚ùå –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ {self.name} –≤ –ë–î...")

    def send_welcome_email(self) -> None:
        """–û—Ç–ø—Ä–∞–≤–∫–∞ –ø–∏—Å—å–º–∞ (—Ä–∞–±–æ—Ç–∞ —Å –ø–æ—á—Ç–æ–π)"""
        print(f"‚ùå –û—Ç–ø—Ä–∞–≤–∫–∞ –ø–∏—Å—å–º–∞ –Ω–∞ {self.email}...")

    def log_user_activity(self, activity: str) -> None:
        """–õ–æ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ (—Ä–∞–±–æ—Ç–∞ —Å –ª–æ–≥–∞–º–∏)"""
        print(f"‚ùå –õ–æ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ: {self.name} - {activity}")


print("‚ùå –ü—Ä–∏–º–µ—Ä –Ω–∞—Ä—É—à–µ–Ω–∏—è SRP:")
user_bad = UserBad("–ò–≤–∞–Ω", "ivan@example.com")
user_bad.validate_email()
user_bad.save_to_database()
user_bad.send_welcome_email()
user_bad.log_user_activity("–∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–ª—Å—è")
````

````
# ‚úÖ –†–ï–®–ï–ù–ò–ï: –∫–∞–∂–¥—ã–π –∫–ª–∞—Å—Å –∏–º–µ–µ—Ç –æ–¥–Ω—É –æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç—å (SRP)

class User:
    """–ü—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è (—Ç–æ–ª—å–∫–æ –¥–∞–Ω–Ω—ã–µ)."""
    def __init__(self, name: str, email: str):
        self.name = name
        self.email = email


class EmailValidator:
    """–û—Ç–≤–µ—á–∞–µ—Ç —Ç–æ–ª—å–∫–æ –∑–∞ –≤–∞–ª–∏–¥–∞—Ü–∏—é email."""
    @staticmethod
    def validate(email: str) -> bool:
        return "@" in email and "." in email.split("@")[-1]


class UserRepository:
    """–û—Ç–≤–µ—á–∞–µ—Ç —Ç–æ–ª—å–∫–æ –∑–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ/–∑–∞–≥—Ä—É–∑–∫—É –∏–∑ –ë–î."""
    @staticmethod
    def save(user: User) -> None:
        print(f"‚úÖ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ {user.name} –≤ –ë–î...")


class EmailService:
    """–û—Ç–≤–µ—á–∞–µ—Ç —Ç–æ–ª—å–∫–æ –∑–∞ –æ—Ç–ø—Ä–∞–≤–∫—É –ø–∏—Å–µ–º."""
    @staticmethod
    def send_welcome_email(user: User) -> None:
        print(f"‚úÖ –û—Ç–ø—Ä–∞–≤–∫–∞ –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Å—å–º–∞ –Ω–∞ {user.email}...")


class Logger:
    """–û—Ç–≤–µ—á–∞–µ—Ç —Ç–æ–ª—å–∫–æ –∑–∞ –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ."""
    @staticmethod
    def log_activity(user_name: str, activity: str) -> None:
        print(f"‚úÖ –õ–æ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ: {user_name} - {activity}")


class UserRegistrationService:
    """–ö–æ–æ—Ä–¥–∏–Ω–∏—Ä—É–µ—Ç –ø—Ä–æ—Ü–µ—Å—Å —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ (–∏—Å–ø–æ–ª—å–∑—É–µ—Ç –≤—Å–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã)."""
    def __init__(self):
        self.validator = EmailValidator()
        self.repository = UserRepository()
        self.email_service = EmailService()
        self.logger = Logger()

    def register_user(self, name: str, email: str) -> bool:
        """–†–µ–≥–∏—Å—Ç—Ä–∏—Ä—É–µ—Ç –Ω–æ–≤–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è."""
        if not self.validator.validate(email):
            print(f"‚ùå Email {email} –Ω–µ–≤–∞–ª–∏–¥–Ω—ã–π!")
            return False

        user = User(name, email)
        self.repository.save(user)
        self.email_service.send_welcome_email(user)
        self.logger.log_activity(name, "–∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–ª—Å—è")
        return True


print("\n‚úÖ –ü—Ä–∏–º–µ—Ä —Å–æ–±–ª—é–¥–µ–Ω–∏—è SRP:")
registration_service = UserRegistrationService()
registration_service.register_user("–ò–≤–∞–Ω", "ivan@example.com")
````

## üîó –ß–∞—Å—Ç—å 8: –ü—Ä–∏–Ω—Ü–∏–ø 13 - –ó–∞–∫–æ–Ω –î–µ–º–µ—Ç—Ä—ã (Law of Demeter)

**Law of Demeter:** "–†–∞–∑–≥–æ–≤–∞—Ä–∏–≤–∞–π—Ç–µ —Ç–æ–ª—å–∫–æ —Å –±–ª–∏–∑–∫–∏–º–∏ –¥—Ä—É–∑—å—è–º–∏!" üë•

- –ú–∏–Ω–∏–º–∏–∑–∏—Ä—É–π—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –º–µ–∂–¥—É –æ–±—ä–µ–∫—Ç–∞–º–∏
- –ò–∑–±–µ–≥–∞–π—Ç–µ —Ü–µ–ø–æ—á–µ–∫ –º–µ—Ç–æ–¥–æ–≤ (`obj.method1().method2().method3()`)
- –£–º–µ–Ω—å—à–∞–µ—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç—å –∏ –ø–æ–≤—ã—à–∞–µ—Ç –≥–∏–±–∫–æ—Å—Ç—å


````
# ‚ùå –ù–ê–†–£–®–ï–ù–ò–ï –ó–∞–∫–æ–Ω–∞ –î–µ–º–µ—Ç—Ä—ã: —Ü–µ–ø–æ—á–∫–∞ –º–µ—Ç–æ–¥–æ–≤

class Engine:
    """–î–≤–∏–≥–∞—Ç–µ–ª—å –∞–≤—Ç–æ–º–æ–±–∏–ª—è."""
    def __init__(self, horsepower: int):
        self.horsepower = horsepower

    def get_horsepower(self) -> int:
        return self.horsepower


class Car:
    """–ê–≤—Ç–æ–º–æ–±–∏–ª—å."""
    def __init__(self, engine: Engine):
        self.engine = engine

    def get_engine(self) -> Engine:
        return self.engine


class Driver:
    """–í–æ–¥–∏—Ç–µ–ª—å."""
    def __init__(self, car: Car):
        self.car = car

    def check_car_horsepower_bad(self) -> int:
        """
        ‚ùå –ü–õ–û–•–û: –æ–±—Ä–∞—â–∞–µ—Ç—Å—è –∫ –Ω–µ–∑–Ω–∞–∫–æ–º—Ü—É —á–µ—Ä–µ–∑ –¥—Ä—É–≥–∞
        Driver -> Car -> Engine (—Ü–µ–ø–æ—á–∫–∞ –º–µ—Ç–æ–¥–æ–≤)
        """
        return self.car.get_engine().get_horsepower()


print("‚ùå –ù–∞—Ä—É—à–µ–Ω–∏–µ –ó–∞–∫–æ–Ω–∞ –î–µ–º–µ—Ç—Ä—ã:")
engine = Engine(200)
car = Car(engine)
driver = Driver(car)
horsepower = driver.check_car_horsepower_bad()
print(f"–ú–æ—â–Ω–æ—Å—Ç—å –¥–≤–∏–≥–∞—Ç–µ–ª—è: {horsepower} –ª.—Å.")
print("‚ö†Ô∏è  –ü—Ä–æ–±–ª–µ–º–∞: Driver –∑–∞–≤–∏—Å–∏—Ç –æ—Ç Car –ò Engine")
````

````
# ‚úÖ –†–ï–®–ï–ù–ò–ï: —Å–æ–±–ª—é–¥–µ–Ω–∏–µ –ó–∞–∫–æ–Ω–∞ –î–µ–º–µ—Ç—Ä—ã

class EngineGood:
    """–î–≤–∏–≥–∞—Ç–µ–ª—å."""
    def __init__(self, horsepower: int):
        self.horsepower = horsepower


class CarGood:
    """–ê–≤—Ç–æ–º–æ–±–∏–ª—å."""
    def __init__(self, engine: EngineGood):
        self.engine = engine

    def get_horsepower(self) -> int:
        """
        ‚úÖ –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥ –≤ Car, —á—Ç–æ–±—ã —Å–∫—Ä—ã—Ç—å –¥–µ—Ç–∞–ª–∏ Engine
        –¢–µ–ø–µ—Ä—å Driver –Ω–µ –Ω—É–∂–Ω–æ –∑–Ω–∞—Ç—å –ø—Ä–æ Engine!
        """
        return self.engine.horsepower


class DriverGood:
    """–í–æ–¥–∏—Ç–µ–ª—å."""
    def __init__(self, car: CarGood):
        self.car = car

    def check_car_horsepower_good(self) -> int:
        """
        ‚úÖ –•–û–†–û–®–û: –æ–±—Ä–∞—â–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –∫ –¥—Ä—É–≥—É (Car)
        Driver -> Car (—Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —É—Ä–æ–≤–µ–Ω—å!)
        """
        return self.car.get_horsepower()


print("\n‚úÖ –°–æ–±–ª—é–¥–µ–Ω–∏–µ –ó–∞–∫–æ–Ω–∞ –î–µ–º–µ—Ç—Ä—ã:")
engine_good = EngineGood(200)
car_good = CarGood(engine_good)
driver_good = DriverGood(car_good)
horsepower = driver_good.check_car_horsepower_good()
print(f"–ú–æ—â–Ω–æ—Å—Ç—å –¥–≤–∏–≥–∞—Ç–µ–ª—è: {horsepower} –ª.—Å.")
print("‚úÖ –ü—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ: Driver –∑–∞–≤–∏—Å–∏—Ç —Ç–æ–ª—å–∫–æ –æ—Ç Car")
print("‚úÖ –ï—Å–ª–∏ Engine –∏–∑–º–µ–Ω–∏—Ç—Å—è, Driver –Ω–µ –ø–æ—Å—Ç—Ä–∞–¥–∞–µ—Ç")
````

## ‚ö†Ô∏è –ß–∞—Å—Ç—å 9: –û–±—Ä–∞–±–æ—Ç–∫–∞ –æ—à–∏–±–æ–∫ –∏ –∏—Å–∫–ª—é—á–µ–Ω–∏–π

**–ü—Ä–∏–Ω—Ü–∏–ø—ã –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏—Å–∫–ª—é—á–µ–Ω–∏–π:**
1. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ `try-except` –¥–ª—è –æ–∂–∏–¥–∞–µ–º—ã—Ö –æ—à–∏–±–æ–∫
2. –õ–æ–≤–∏—Ç–µ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–µ –∏—Å–∫–ª—é—á–µ–Ω–∏—è, –Ω–µ –≤—Å–µ –ø–æ–¥—Ä—è–¥
3. –ü—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è–π—Ç–µ –ø–æ–ª–µ–∑–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± –æ—à–∏–±–∫–µ
4. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ `finally` –¥–ª—è –æ—á–∏—Å—Ç–∫–∏ —Ä–µ—Å—É—Ä—Å–æ–≤
5. –°–æ–∑–¥–∞–≤–∞–π—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∏—Å–∫–ª—é—á–µ–Ω–∏—è –¥–ª—è —Å–ø–µ—Ü–∏—Ñ–∏—á–Ω—ã—Ö –æ—à–∏–±–æ–∫


````
# ‚ùå –ü–õ–û–•–û–ô –ü–†–ò–ú–ï–†: –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–π

def divide_bad(a: float, b: float) -> float:
    """–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –æ—à–∏–±–∫–∏."""
    try:
        return a / b
    except:  # ‚ùå –õ–æ–≤–∏—Ç –í–°–ï –∏—Å–∫–ª—é—á–µ–Ω–∏—è - —ç—Ç–æ –ø–ª–æ—Ö–∞—è –ø—Ä–∞–∫—Ç–∏–∫–∞!
        print("–û—à–∏–±–∫–∞")
        return 0


print("‚ùå –ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞:")
result = divide_bad(10, 0)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {result}")
````

````
# ‚úÖ –•–û–†–û–®–ò–ô –ü–†–ò–ú–ï–†: –ø—Ä–∞–≤–∏–ª—å–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–π

class DivisionError(Exception):
    """–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–µ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ –¥–ª—è –æ—à–∏–±–æ–∫ –¥–µ–ª–µ–Ω–∏—è."""
    pass


def divide_good(a: float, b: float) -> float:
    """
    –î–µ–ª–∏—Ç a –Ω–∞ b —Å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –∏—Å–∫–ª—é—á–µ–Ω–∏–π.

    Args:
        a: –î–µ–ª–∏–º–æ–µ
        b: –î–µ–ª–∏—Ç–µ–ª—å

    Returns:
        –†–µ–∑—É–ª—å—Ç–∞—Ç –¥–µ–ª–µ–Ω–∏—è

    Raises:
        TypeError: –ï—Å–ª–∏ –∞—Ä–≥—É–º–µ–Ω—Ç—ã –Ω–µ —á–∏—Å–ª–∞
        DivisionError: –ï—Å–ª–∏ –ø–æ–ø—ã—Ç–∫–∞ –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ–ª—å
    """
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ç–∏–ø–æ–≤
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("–û–±–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —á–∏—Å–ª–∞–º–∏")

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ–ª—å
    if b == 0:
        raise DivisionError("–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ!")

    return a / b


print("\n‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞:")

# –¢–µ—Å—Ç 1: –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ
try:
    result = divide_good(10, 2)
    print(f"10 / 2 = {result}")
except (TypeError, DivisionError) as e:
    print(f"‚ùå –û—à–∏–±–∫–∞: {e}")

# –¢–µ—Å—Ç 2: –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å
try:
    result = divide_good(10, 0)
    print(f"10 / 0 = {result}")
except (TypeError, DivisionError) as e:
    print(f"‚ùå –û—à–∏–±–∫–∞: {e}")

# –¢–µ—Å—Ç 3: –Ω–µ–≤–µ—Ä–Ω—ã–π —Ç–∏–ø
try:
    result = divide_good(10, "–¥–≤–∞")
    print(f"10 / '–¥–≤–∞' = {result}")
except (TypeError, DivisionError) as e:
    print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
````
