# ========== РАЗДЕЛ 3: [ГЕНЕРАТОРЫ, ДЕКОРАТОРЫ](https://docs.google.com/presentation/d/1fVHnDg4ExFJAQfyQ7fBuLn_XbvP50Hk56hb5nnXo1yY/edit?usp=sharing), *args/**kwargs ==========

```python
print("\n\n" + "="*60)
print("РАЗДЕЛ 3: ГЕНЕРАТОРЫ И ДЕКОРАТОРЫ")
print("="*60)

# 1. List Comprehensions
print("\n1. LIST COMPREHENSIONS")
print("-"*60)

# Простая list comprehension
squares = [x**2 for x in range(1, 6)]
print(f"Квадраты: {squares}")

# С условием
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print(f"Квадраты чётных: {even_squares}")

# с несколькими циклами
matrix = [[i+j for j in range(3)] for i in range(3)]
print(f"Матрица:\n{np.array(matrix)}")

# 2. Генераторы
print("\n2. ГЕНЕРАТОРЫ (yield)")
print("-"*60)

def fibonacci_gen(n):
    """Generator for Fibonacci numbers"""
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print("Первые 10 чисел Fibonacci:")
fib_list = list(fibonacci_gen(10))
print(fib_list)

# 3. *args и **kwargs
print("\n3. *args и **kwargs")
print("-"*60)

def process_data(*args, **kwargs):
    """Process variable arguments and keyword arguments"""
    print(f"Позиционные аргументы (*args): {args}")
    print(f"Ключевые аргументы (**kwargs): {kwargs}")

process_data(1, 2, 3, name="Alice", age=30, city="NYC")

# 4. Декораторы
print("\n4. ДЕКОРАТОРЫ")
print("-"*60)

def timer_decorator(func):
    """Decorator to measure execution time"""
    def wrapper(*args, **kwargs):
        import time
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Время выполнения: {end - start:.4f} сек.")
        return result
    return wrapper

@timer_decorator
def slow_function():
    """A slow function for demonstration"""
    import time
    time.sleep(0.1)
    return "Done!"

result = slow_function()
print(f"Результат: {result}")
```

# 3. ОПЕРАТОРЫ: Арифметические, сравнения, логические

```python
print("\n3. ОПЕРАТОРЫ")
print("-"*60)

# Арифметические операторы
a, b = 15, 4
print(f"\nАрифметические операторы (a={a}, b={b}):")
print(f"  a + b = {a + b}  # сложение")
print(f"  a - b = {a - b}  # вычитание")
print(f"  a * b = {a * b}  # умножение")
print(f"  a / b = {a / b}  # деление")
print(f"  a // b = {a // b}  # целочисленное деление")
print(f"  a % b = {a % b}  # остаток")
print(f"  a ** b = {a ** b}  # возведение в степень")

# Операторы сравнения
print(f"\nОператоры сравнения:")
print(f"  {a} > {b} => {a > b}")
print(f"  {a} < {b} => {a < b}")
print(f"  {a} == {b} => {a == b}")
print(f"  {a} != {b} => {a != b}")
print(f"  {a} >= {b} => {a >= b}")
print(f"  {a} <= {b} => {a <= b}")

# Логические операторы
cond1, cond2 = True, False
print(f"\nЛогические операторы (cond1={cond1}, cond2={cond2}):")
print(f"  cond1 and cond2 => {cond1 and cond2}")
print(f"  cond1 or cond2 => {cond1 or cond2}")
print(f"  not cond1 => {not cond1}")
print(f"  not cond2 => {not cond2}")
```

# ========== РАЗДЕЛ 2: ФУНКЦИИ И УПРАВЛЕНИЕ ПОТОКОМ ==========

```python
print("\n\n" + "="*60)
print("РАЗДЕЛ 2: ФУНКЦИИ И УПРАВЛЕНИЕ ПОТОКОМ")
print("="*60)

# 1. Основные функции
print("\n1. Основные функции")
print("-"*60)

# Простая функция
def calculate_average(numbers):
    """
    Калькулирует среднее арифметическое данных.
    """
    return sum(numbers) / len(numbers)

data = [10, 20, 30, 40, 50]
avg = calculate_average(data)
print(f"Среднее значение: {avg}")

# Функция с вызванным параметром
def greet(name, age=25):
    """
    Приветствует по имени и возрасту.
    """
    return f"Hello, {name}! You are {age} years old."

print(f"\n{greet('Alice')}")
print(f"{greet('Bob', 30)}")

# Функция с несколькими выходными результатами
def calculate_stats(numbers):
    """
    Возвращает минимум, максимум, среднее.
    """
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

min_val, max_val, avg_val = calculate_stats([5, 15, 25, 35, 45])
print(f"\nМин: {min_val}, Макс: {max_val}, Сред: {avg_val:.2f}")
```

# 2. УПРАВЛЕНИЕ ПОТОКОМ: if/else/elif, for, while

```python
print("\n2. УПРАВЛЕНИЕ ПОТОКОМ")
print("-"*60)

# if/else/elif обычные примеры
print("\nif/else/elif:")
for score in [45, 55, 75, 85, 95]:
    if score < 50:
        grade = "F"
    elif score < 60:
        grade = "D"
    elif score < 70:
        grade = "C"
    elif score < 80:
        grade = "B"
    else:
        grade = "A"
    print(f"  Score: {score} => Grade: {grade}")

# for цикл
print("\nfor loop:")
print("  Натуральные числа от 1 до 10:")
for i in range(1, 11):
    print(f"    {i}", end=" ")
print()

# for с энумерацией
print("\n  Обработка списка:")
for idx, name in enumerate(["Alice", "Bob", "Charlie"], 1):
    print(f"    {idx}. {name}")

# while цикл
print("\nwhile loop:")
count = 0
while count < 5:
    print(f"  Count: {count}")
    count += 1

# Цикл с break/continue
print("\nfor loop с break/continue:")
for i in range(10):
    if i == 3:
        continue  # Пропускаем 3
    if i == 7:
        break  # Останавливаем при 7
    print(f"  {i}", end=" ")
print()
```
